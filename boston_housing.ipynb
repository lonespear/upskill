**Linear Regression Demonstration**

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/lonespear/upskill/refs/heads/main/Boston.csv')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['MEDV'], axis=1), df['MEDV'], test_size=0.2, random_state=42)
model_lr = LinearRegression().fit(X_train, y_train)
model_lr.score(X_test, y_test)
```

```python
mean_squared_error(y_test, model_lr.predict(X_test))
```

```python
import statsmodels.api as sm
import numpy as np

model = sm.OLS(y_train,X_train)
results = model.fit()
results.summary()
```

```python
# Make predictions
y_pred = model.predict(X_test)

from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```
